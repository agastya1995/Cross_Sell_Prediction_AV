```python
X['Policy_Sales_Channel'] = X['Policy_Sales_Channel'].astype('int64')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from xgboost import XGBClassifier as xgbclf
import xgboost as xgb
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.metrics import confusion_matrix
from bayes_opt import BayesianOptimization
from sklearn.metrics import auc
```

```python
xgb_preds = xgb.predict(X_test)
```

```python
xgb_preds = xgb_preds.astype(int)
```

```python
y_test = y_test.astype('int')
```

```python
print(classification_report(y_test, xgb_preds))
```

```python
print(roc_auc_score(y_test, xgb_preds))
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, xgb_preds).ravel()
(tn, fp, fn, tp)
```

```python
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, xgb_preds)
roc_auc = auc(false_positive_rate, true_positive_rate)
```

```python
plt.title('ROC CURVE')
plt.plot(false_positive_rate, true_positive_rate, 'b',
label='AUC = %0.2f'% roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1],'r--')
plt.xlim([-0.1,1.2])
plt.ylim([-0.1,1.2])
plt.ylabel('True_Positive_Rate')
plt.xlabel('False_Positive_Rate')
plt.show()
```
